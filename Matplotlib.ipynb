
## Short Format Notation

---


matplotlib proivdes a short format notation for easily providing the line plot parameters. They are defined as follows

`fmt = '[marker][line][color]'`


```
'b'    # blue markers with default shape
'or'   # red circles
'-g'   # green solid line
'--'   # dashed line with default color
'^k:'  # black triangle_up markers connected by a dotted line
```




````python
Temperaturebangalore = [36,32,30,25,28,32,34,31,30,33]
TemperatureDelhi =     [40,39,31,35,39,37,36,35,34,38]
plt.figure(figsize=(6,4),dpi=100)
plt.plot(Temperaturebangalore,'^k:',label="Bengaluru")
plt.plot(TemperatureDelhi,'ob-',label="Delhi")
plt.legend()
plt.title("Temperature Bengaluru vs Delhi")
plt.show()
````

To use latex expressions like $f(x) = y^2$ or $10e^{-5}$  within the plots, we need to first activate the rcParams for matplotlib related to `text.usetex`

```

```

### Excersise 2

Clean up the histogram such that only top 5 man of the match candidates are shown and they are in sorted order. Do not use pandas library functions

Expected output

````python
#@title Solution for Excersise 2

## dict (hashmap) approach Finishes in O(N), a count and loop method may take O(N^2) on worst case
count  = {}
for player in manofthematch:
  if player in count:
    count[player] += 1
  else:
    count[player] = 1

## now, we need to sort the dict based on keys
## [Ref link](https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value)
newCount = {k: v for k, v in sorted(count.items(),reverse=True, key=lambda item: item[1])}

## Obtain First Five Values and plot
Player = list(newCount.keys())
mom = list(newCount.values())

plt.figure(figsize=(6,4),dpi=100)
# print(Player)
# print(mom)

plt.bar(Player[:5],mom[:5], color='r')

plt.show()
````

````python
#@title Solution for Gaussian

## get a normal distribution from numpy
## coded by : Divij Ghose
from scipy import stats

x_data = np.arange(-3, 3, 0.001)
y_data = stats.norm.pdf(x_data, 0, 1)
plt.figure(figsize=(6,4),dpi=100)
plt.plot(x_data,y_data,color='k')
colorlist = ['#F0F8FF','#98F5FF','#00688B','#104E8B']
xlabels = [r'$-3\sigma$',r'$-2\sigma$',r'$-1\sigma$',r'$\mu$',r'$1\sigma$',r'$2\sigma$',r'$3\sigma$']
for i in range(4):
    plt.fill_between(x_data[(i-1)*1000:i*1000],y_data[(i-1)*1000:i*1000],color=colorlist[i],alpha=i*0.25)
    plt.fill_between(x_data[-i*1000:(i-1)*-1000-1],y_data[-i*1000:(i-1)*-1000-1],color=colorlist[i],alpha=i*0.25)
    plt.axvline(x=x_data[i*1000],color='w')
    plt.axvline(x=x_data[-(i*1000)],color='w')

plt.xticks(np.arange(-3,4,1),xlabels)
plt.text(0.1, 0.25, '34.1%',color='w')
plt.text(-0.75, 0.25, '34.1%',color='w')
plt.text(1.1, 0.01, '13.6%',color='w')
plt.text(-1.75,.01, '13.6%',color='w')
plt.text(2.2, 0.05, '2.1%',color='k')
plt.text(-2.8,.05, '2.1%',color='k')
plt.title("Gaussian Distribution with 0-Mean and 1-Std.Dev")

````


# Difference between plt and ax,fig objects

---

**The Major Challenge** , that comes to people when they start learning matplotlib is understanding the difference between the `plt` object and the `figure` and `axes` object


What ever, we have learnt untill now, we have directly used the `plt` objects method to plot the necessary plots. Something like the one below


```
plt.figure(figsize=(6,4),dpi=100)
plt.plot(x, y, color='r')
```

However, if you ever refer to any tutorials otr help on stackoverflow, you will be comming across these syntaxes, which might look like

```
fig,ax = plt.subplots(2,2,figsize(6,4))  ## getting axes and Figure object
ax.plot(x, y)                            ## plot using axes object
fig.saveplt()                            ## Save Plot using the figure object
```

or

```
figure = plt.figure()                 ## Get a figure object
new_plot = figure.add_subplot(111)    ## Add an axes object onto the figure
new_plot.plot(x, y)                   ## plot using the axes object
```



**So, What is happening here?**


It seems that the `plt` object is the wrapper that was made on top of existing matplotlib libraries to make sure the plotting syntaxes remain similar to the matlab's plotting syntax [Reference](https://stackoverflow.com/questions/37970424/what-is-the-difference-between-drawing-plots-using-plot-axes-or-figure-in-matpl)



Lets try to understand what data these objects hold in general


### 1. Figure Object

This object acts like a `canvas` on which the plotting will be done. So the details such as size of the figure, number of subplots a figure has and saving/ exporting the figure will be contributed by this object.

The figure object can be initiated using `plt.figure()` or using the `plt.subplots()` or `plt.subplot()` object

````python
## obtain a figure object
fig = plt.figure()
````

Note : The Figure object will not be sufficient to add plots, we need axes object to make the actual plots within the canvas


### 2. Axes Object

The Axes object interacts with the underlying graphical libraries to generate the basic plotting objects such as a line2d, circle , etc. these artifacts are drawn based on the inputs given by the user and then placed on a canvas generated by the figure object.

The Axes object has menthods like `plot()`, `bar()` , `scatter()` etc and also the helper functions like `set_title()`, `set_xlabel()` , `set_ylabel()`, `set_legend()` etc


A axes object can be created and added to a figure object by either one of the below syntax


````python
## Create an axes and figure object directly
fig, axes = plt.subplots()

## OR
fig = plt.figure()
ax = fig.add_subplot()
````

### 3. Plt Object

The plot object is the wrapper to both the axes and the figure object. When we create the plot object both axes object and the figure object are created and will be stored inside the internal datastructure of plt.


These objects can be extracted from the plt objects using the command
` plt.gcf()` for figure and `plt.gca()` for axes


**Hint**

Most of the (not all) methods that are available with `plt` such as `title` , `xlabel` , `ylabel` will be available for axes as `set_title`, `set_xlabel`,`set_ylabel`



## Creating subplots ( Introduction )

We can create subplots by using the `subplots()` function. This function will return a grid of axes object which we can use to plot the subplots.




````python
#### FOUR DIFFERENT SUBPLOTS
import numpy as np
fig, ax = plt.subplots(2,2,figsize=(11,8))

#Subplot -1 - Bar plot
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax[0,0].bar(langs,students)
ax[0,0].set_title("Bar Plot")
ax[0,0].set_xlabel("Test label")

#Subplot -2 - histogram
x = np.random.normal(170, 10, 250)
ax[0,1].hist(x,color="green")
ax[0,1].set_title("Histogram Plot")


# Subplot -3 , scatter
girls_grades = [89, 90, 70, 89, 100, 80, 90, 100, 80, 34]
boys_grades = [30, 29, 49, 48, 100, 48, 38, 45, 20, 30]

ax[1,0].scatter(grades_range, girls_grades, color='r')
ax[1,0].scatter(grades_range, boys_grades, color='b')
ax[1,0].set_xlabel('Grades Range')
ax[1,0].set_ylabel('Grades Scored')
ax[1,0].set_title('scatter plot')

# Subplot -4 , Contour
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
cp = ax[1,1].contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax[1,1].set_title('Filled Contours Plot')
#ax.set_xlabel('x (cm)')
ax[1,1].set_ylabel('y (cm)')

# plt.subplots_adjust(hspace=0.5)
fig.tight_layout()

fig.suptitle("Main title")
````

#### Tight Layout

This function call ensures that all the elements within the figure are properly spaced and there is no overlap between each subplot and ensures every object resides within the main canvas


The `tight_layout()` object also accepts attributes such as `pad, w_pad and h_pad.`, which can further control the spacing between subplots.


Note : `plt` object has `subplots_adjust()` method which can add padding between subplots

We can generate any kinds of slightly complex plots such as combined bar chart using the `fig` and `axes` object.

````python
import numpy as np
import matplotlib.pyplot as plt

N = 3
ind = np.arange(N)  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure()
ax = fig.add_subplot(111)

yvals = [4, 9, 2]
rects1 = ax.bar(ind, yvals, width, color='r')
zvals = [1,2,3]
rects2 = ax.bar(ind+(width/2), zvals, width, color='g')
kvals = [11,12,13]
rects3 = ax.bar(ind+width, kvals, width, color='b')

ax.set_ylabel('Scores')
ax.set_xticks(ind+width)
ax.set_xticklabels( ('2011-Jan-4', '2011-Jan-5', '2011-Jan-6') )
ax.legend( (rects1[0], rects2[0], rects3[0]), ('y', 'z', 'k') )

def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.05*h, '%d'%int(h),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

plt.show()
````

## Animations using Matplotlib

---


The Animation classes in matplotlib can be accomplished using these classes

image.png


* Artistic Animation -- Uses the Artist objects to create the animation
* Function Animation -- uses function, which will be called repeatitively to generate animations



[Matplotlib Animation](Reference)



### Method - 1 : Render Each Frame Seperately
---

In this, we will use the `pause()` function within matplotlib to generate a plot, which will make it appear like an animation



````python
y = ecg[300:4000:10]
t = time[300:4000:10]

## generate Plot Object
fig,ax = plt.subplots(figsize=(6,4),tight_layout=True,dpi=100 )

for frame in range(len(y)):
    ax.clear()  ## Clear the Object
    ax.set_xlabel("Time(s)")
    ax.set_ylabel("ECG in mV")
    ax.grid()
    ax.plot(t[:frame],y[:frame],'ok-',markevery=[-1])
    ax.set_ylim([-1,1.5])
    fig.canvas.draw()
    plt.pause(0.00001)
    pass;
````

### Method - 2 : Update the Line Elements(Artists Object)
---


In this, Rather than redrawing the entire plot, we can just modify the data which is stored in the line object to get the updated plots.






````python
y = ecg[300:4000:10]
t = time[300:4000:10]

## generate Plot Object
fig,ax = plt.subplots(figsize=(6,4),tight_layout=True,dpi=100 )

ln, = ax.plot([],[],'ok-',markevery=[-1]) ### Obtain the Line Object
ax.set_xlabel("Time(s)")
ax.set_ylabel("ECG in mV")

for frame in range(len(y)):


    ln.set_data(t[:frame],y[:frame])
    ax.grid()
    ax.plot(t[:frame],y[:frame],'ok-',markevery=[-1])   ## rather than clearing the canvas, just update the line object alone.
    # ax.set_ylim([-1,1.5])
    ax.relim()
    fig.canvas.draw()
    plt.pause(0.00001)
    pass;
````

## 3. Using FuncAnimation and blit
---

This is matplotlib's inbuilt funcitonality to animate the plots by calling this functions again and again


* `init()`   -- Function initialises the plot
* `update()` -- Function updates the artists objects to update the plot

**blit** is an intelligent way to update the graphics without much effort. Additional details about blit can be refered from this [video](https://www.youtube.com/watch?v=-6g1tD66QBA)

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import mp

y = ecg[300:10000:10]
t = time[300:10000:10]


fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [], 'ok-')


## Intialises the plot
def init():
    ln, = ax.plot([],[],'ok-',markevery=[-1]) ### Obtain the Line Object
    ax.set_xlabel("Time(s)")
    ax.set_ylabel("ECG in mV")
    ax.set_ylim([-1,1.5])
    return ln,

## Updates the plot at each frame
def update(data):
    xdata.append(data[1])
    ydata.append(data[0])
    ln.set_data(xdata, ydata)
    ax.relim()
    return ln,

ani = FuncAnimation(fig, update, frames=list(zip(y,t)),
                    init_func=init, interval = 10, blit=True)
plt.show()
````
