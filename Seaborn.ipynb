## 1. Installation
---

This package can be installed using python-pip

`pip install seaborn`

```python
# Import seaborn
import seaborn as sns

iris  = sns.load_dataset("iris")
```

## 2. Distribution Plot
---

Commonly used parameters

* kde - Kernel Density Estimator
* bins - No of Histogram Bins

```python
iris
```

```python
sns.displot(iris["sepal_length"],kde=True)
```

## 3. Joint Plot
---

plot the joint distribution between two attributes in a given plot. Can provide distribution and co-relation between the two parameters



```python
## load Car Crash data
car = sns.load_dataset("car_crashes")

sns.jointplot(x="speeding",y="alcohol",data=car,kind='reg')
# sns.jointplot(car["speeding"],car["alcohol"],kind='reg')
```

## 4. KDE Plot
---

Plots the KDE ( generaly available within the displot function ) alone

```python
sns.kdeplot(car["speeding"])
```

## 5. Pair Plot
---

Gives the relation between all the variables ( columns ) avaiable within the dataframe

This is very usefull for performing EDA and identifying corelation between the parameters

Note: For Larger datasets, it will take longer time to generate the datasets


```python
sns.pairplot(iris)
```

## 6. Styling in Seaborn
---

seaborn by default has lot of inbuilt styling functionalities. They can be changed using the `set_style` function

[Reference](https://seaborn.pydata.org/generated/seaborn.set_style.html)

* Color maps can be refered from [cmap Reference](https://matplotlib.org/stable/tutorials/colors/colormaps.html)

```python
sns.set_style("whitegrid")
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])
```

```python
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.lineplot(x=["A", "B", "C"], y=[1, 3, 2])
```

## 7. Box plot
---

Provides the range of the data and the extreme values within the given data

```python
sns.boxplot(x = "species" ,y="sepal_length",data=iris)
```

```python
import numpy as np
import pandas as pd
a = np.random.randint(0,10,50)
a = list(a)
a.append(100000)
df = pd.DataFrame(a)
df
```

```python
sns.boxplot(y=0,data=df)
```

```python
## With hue
df = sns.load_dataset("penguins")
sns.boxplot(data=df, x="island", y="body_mass_g",hue="sex")
```

## 8. Count Plot
---

Displays the count of the data in given column

```python
tips = sns.load_dataset("tips")
sns.countplot(x="sex",data=tips)
```

## 9. Violin Plot
---

Similar to box plot, however, it also provides the distribution of data

Note : These plots are little insensitive to the outliers in the data, since they are plotted seperately and not included while calculating/plotting the distribution

```python
sns.violinplot(data=df, x="island", y="body_mass_g", hue="sex")
```

#### Swarm plot
---

Similar to violin plot but in discrete points

```python
sns.swarmplot(data=df, x="island", y="body_mass_g",color='black')
sns.violinplot(data=df, x="island", y="body_mass_g")
```

10. Heat Map
---

generates a heat map from given matrix data


```python
car_correlation = car.corr()
sns.heatmap(car_correlation,annot=True,cmap='Blues')
```
